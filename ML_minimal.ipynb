```python
df = pd.read_csv('gdrive/MyDrive/colabFolderCS410/Mall_Customers.csv')
df
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split


x = df.loc[:, df.columns != 'Spending Score (1 - 100)']
y = df['Spending Score (1-100)']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
import lazypredict

# importing regression class from lazypredict
from lazypredict.Supervised import LazyRegressor

# importing classification class from lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
# Defining our model using lazyregressor
multiple_ML_model = LazyRegressor(verbose=0, ignore_warnings=True, predictions=True)

# fitting the models along with predicting with outputs of each model
models, predictions = multiple_ML_model.fit(x_train, x_test, y_train, y_test)
```

```python
models
```

```python
predictions
```

```python
# defining our model using lazyclassifier
multiple_ML_model_2 = LazyClassifier(verbose=0, ignore_warnings=True, predictions=True)

# fitting the model along with predicting with outputs of each model
models, predictions = multiple_ML_model_2.fit(x_train, x_test, y_train, y_test)
```
